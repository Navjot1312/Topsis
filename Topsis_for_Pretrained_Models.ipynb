```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'Precision': Precision, 'Recall': Recall, 'F1-score': F1_score})
df.index = pd.RangeIndex(start=1, stop=len(df) + 1, name='Model')
```

```python
print(df)
```

```python
weights=[1,1,1]
impact=[1,1,1]
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
norm_data=df/np.sqrt((df ** 2).sum(axis=0))
```

```python
norm_data=norm_data*weights

rough=norm_data*impact
```

```python
best=rough.max().abs()

worst=rough.min().abs()
```

```python
dist_best=np.sqrt(((norm_data-best)**2).sum(axis=1))

dist_worst=np.sqrt(((norm_data-worst)**2).sum(axis=1))
```

```python
total_dist=dist_best+dist_worst
performance=dist_worst/total_dist
rank = pd.Series(performance, name='Performance').rank(ascending=False).astype(int)
```

```python
norm_data['Topsis Score']=performance
norm_data['Rank']=rank
```
